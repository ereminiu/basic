```python
counter = Counter(df[df.Gender == 'M'].Age)
h, labels = [counter[x] for x in counter], [x for x in counter]
h.sort()
labels.sort()
```

```python
x = np.arange(7)

cols = ['b', 'g', 'r', 'c', 'm', 'y', 'purple']
for i in range(7):
    plt.bar(x[i], h[i], color=cols[i], label=labels[i])

plt.ylabel('Users Distibution by Age')
plt.xlabel('Age group')
plt.legend()
# plt.savefig('Male_only.jpg')
plt.show()
```

```python
plt.pie(h, labels=labels, colors=cols)

plt.show()
```
