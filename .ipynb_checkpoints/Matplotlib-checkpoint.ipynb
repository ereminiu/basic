```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = np.arange(1, 100)
plt.plot(x, x**2, color='r', label='O($n^{2}$)')
plt.plot(x, x*np.log2(x), color='b', label='O($n*\log_2{n}$)')
plt.plot(x, x*np.sqrt(x), color='g', label='O($n*\sqrt{n}$)')
plt.plot(x, x, color='m', label='O(n)')
plt.axis([0, 100, 0, 1000])
plt.ylabel('time', fontsize=15)
plt.xlabel('n', fontsize=15)
plt.legend(loc='best', fontsize=8)
plt.savefig('asimptotic_compatation.jpg')
plt.show()
```

```python
x = np.linspace(0, 10, 5)

valera = np.random.poisson(5, 5)
karina = np.random.poisson(5, 5)

plt.plot(x, valera, 'r', label='valera')
plt.plot(x, karina, 'b', label='karina')

plt.legend()
plt.savefig('valera_distribution.jpg')
plt.show()
```

```python
x = np.linspace(0, 10, 5)

karina = [2, 5, 1, 8, 3]
plt.bar(x, karina, alpha=0.9, color='r')
valera = [1, 2, 3, 5, 7]
plt.bar(x, valera, alpha=0.8)

plt.show()
```
